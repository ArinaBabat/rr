```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('titanic.csv')
df['Relatives']=df['SibSp']+df['Parch']
df
```

```python
df.loc[(df.Pclass==1),'Pclass']='Элита'
df.loc[(df.Pclass==2),'Pclass']='Средний класс'
df.loc[(df.Pclass==3),'Pclass']='Работяги'
df
```

```python
q=df['Pclass'].value_counts()
q
```

```python
df['Price']=df['Fare']
df.loc[(df.Price<20),'Price']='Дёшево'
df.loc[(df.Price!='Дёшево'),'Price']='Дорого'
df
```

```python
fdf=df.sort_values(by='Fare',ascending=False)
fdf
```

```python
x=df['SibSp']
y=df['Parch']
x=np.array(x)
y=np.array(y)
x=x+y
df['FamNum']=x
count=['Survived','Pclass','Age','SibSp','Parch','Fare','FamNum']
sns.boxplot(y='Age',x='Pclass',data=df)
```

```python
fd=[]
fa=[]
md=[]
ma=[]
for i in range(df.shape[0]):
  if df['Survived'][i]==0:
    if df['Sex'][i]=='male':
      md.append(df['Age'][i])
    else:
      fd.append(df['Age'][i])
  else:
    if df['Sex'][i]=='male':
      ma.append(df['Age'][i])
    else:
      fa.append(df['Age'][i])
ma=pd.DataFrame(ma)
md=pd.DataFrame(md)
fa=pd.DataFrame(fa)
fd=pd.DataFrame(fd)
gr=np.arange(0, 100, 3)
pd.cut(fd[0],bins=gr).value_counts(sort=False).plot.line(color='purple',label='Погибшие женщины')
pd.cut(fa[0],bins=gr).value_counts(sort=False).plot.line(color='plum',label='Выжившие женщины')
pd.cut(md[0],bins=gr).value_counts(sort=False).plot.line(color='navy',label='Погибшие мужчины')
pd.cut(ma[0],bins=gr).value_counts(sort=False).plot.line(color='skyblue',label='Выжившие мужчины')
plt.legend()
```
